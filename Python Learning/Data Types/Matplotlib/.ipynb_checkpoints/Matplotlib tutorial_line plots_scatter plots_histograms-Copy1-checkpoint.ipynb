```python
import matplotlib.pyplot as plt
%matplotlib inline

x = list(range(0, 10))
y = list(range(-10, 0))

plt.plot(x, y)
```

```python
a = [0, -100, 25, 67, -323]
b = [0, 3, 7, 3, 9]
plt.plot(a, b)
```

```python
#focus in on the triangle above. x axis starts at -50 and ends around 80. and y axis starts at 2, and ends around 8. 
plt.axis([-50, 80, 2, 8 ])
plt.plot(a, b)
```

```python
#adding labels
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8 ])
#replaces the x-axis variables with letters. e.g"helloo"
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', 'o'))
#makes color red, by adding color="red"
plt.plot(a, b, color="red")
```

```python
#different kind of graph, histogram.
plt.hist(a)
```

```python
#histogram
plt.hist(x)
plt.hist(y)
```

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

```python
#scatter plot
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, pop)
plt.show()
```

```python
#divided the histogram by specifying bins=3
values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(values, bins=3)
plt.show()
```
