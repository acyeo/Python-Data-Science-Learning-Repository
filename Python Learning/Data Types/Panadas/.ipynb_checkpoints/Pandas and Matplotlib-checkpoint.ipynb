```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
#inserting a legend
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
#using real data example
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
#compare the population growth in the US and China. isloate the data for US and China
```

```python
us = data[data.country == 'United States']
china = data[data.country == 'China']
```

```python

china
```

```python
us
```

```python
us.population
```

```python
us.population.iloc[0]
```

```python
#this gives us the formulated output in percent growth by dividing all values by the first US population or [0] and multiplying by 100
us.population / us.population.iloc[0] * 100
```

```python
#how to show % growth rate
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(["US", "China"])
plt.xlabel('year')
plt.ylabel("population (%)")
plt.show()
```
